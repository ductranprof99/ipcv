```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import NoNorm
```

```python
def show_image(image, title="sample", figsize=(8,6), tickoff=False):
    plt.figure(figsize=figsize)
    if image.ndim == 2:
        plt.imshow(image, cmap="gray", norm=NoNorm())
    else:
        plt.imshow(image)
    plt.title(title)
    plt.show()

def show_double_image(image1,image2, title="Sample",figsize=(8,6)):
    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=figsize)
    fig.suptitle(title)
    ax1.imshow(image1,cmap="gray", norm=NoNorm())
    ax2.imshow(image2,cmap="gray", norm=NoNorm())
```

```python
img = cv2.imread('./images/kids.png',0)
img = cv2.medianBlur(img,5)
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding','Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
